
````python
import threading
import time

def download(file_name):
    print(f"Starting download: {file_name}")
    time.sleep(2)  
    print(f"Finished download: {file_name}")

files = ["file1", "file2", "file3", "file4", "file5"]

threads = []
for file_name in files:
    t = threading.Thread(target=download, args=(file_name,))
    threads.append(t)
    t.start()   

for t in threads:
    t.join()

print("All downloads completed")
````

````python
from multiprocessing import Process
import time

def compute_square(number):
    print(f"Computing square of {number}")
    time.sleep(1) 
    print(f"Square of {number} is {number * number}")

numbers = list(range(1, 9))

if __name__ == "__main__":
    processes = []
    for number in numbers:
        p = Process(target=compute_square, args=(number,))
        processes.append(p)
        p.start()  

    for p in processes:
        p.join()

    print("All computations completed")
````

## **Problem 3: Asynchronous Program**
Create an asynchronous program with three tasks:

````python
import asyncio
import nest_asyncio

async def load_data():
    print("load_data started")
    await asyncio.sleep(2)  
    print("load_data finished")

async def process_data():
    print("process_data started")
    await asyncio.sleep(2)
    print("process_data finished")

async def save_data():
    print("save_data started")
    await asyncio.sleep(2)
    print("save_data finished")

async def main():
    await asyncio.gather(
        load_data(),
        process_data(),
        save_data(),
    )
    print("All tasks completed")

nest_asyncio.apply()
asyncio.run(main())
````

## **Problem 1: Comparing Threading and Multiprocessing** 
Create two versions of the same program:
- a threading version

- a multiprocessing version

The program **should**:

- compute factorial values

- simulate heavy computation using `time.sleep`

Numbers to process:

`10, 11, 12, 13`

**Tasks**:

- measure execution time for each version

- determine which approach is faster
----

````python
import threading
import multiprocessing as mp
import math
import time

NUMBERS = [10, 11, 12, 13]

def heavy_factorial(n):
    time.sleep(1)          # simulate heavy computation
    return math.factorial(n)

def threading_version(numbers):
    results = {}
    lock = threading.Lock()

    def worker(n):
        r = heavy_factorial(n)
        with lock:
            results[n] = r

    threads = [threading.Thread(target=worker, args=(n,)) for n in numbers]
    for t in threads: t.start()
    for t in threads: t.join()
    return results

def mp_version(numbers):
    with mp.Pool(processes=len(numbers)) as pool:
        results = pool.map(heavy_factorial, numbers)
    return dict(zip(numbers, results))

if __name__ == "__main__":
    start = time.perf_counter()
    t_res = threading_version(NUMBERS)
    t_time = time.perf_counter() - start
    print("Threading results:", t_res)
    print(f"Threading time: {t_time:.2f}s\n")

    start = time.perf_counter()
    m_res = mp_version(NUMBERS)
    m_time = time.perf_counter() - start
    print("Multiprocessing results:", m_res)
    print(f"Multiprocessing time: {m_time:.2f}s\n")

    print("Faster approach:", "Threading" if t_time < m_time else "Multiprocessing")
````

## **Problem 2: Simple Task Scheduler**
Create a basic task scheduler.

**Requirements**:

- Tasks should execute after a specified delay

- Store tasks in a queue

- Workers should periodically check the queue

- Execute tasks when the scheduled time arrives

**Hints**:

- use time.time()

- store (execution_time, task) pairs in the queue
---

````python
import threading
import queue
import time

class TaskScheduler:
    def __init__(self, poll_interval=0.1):
        self.q = queue.PriorityQueue()   # ordered by execution_time
        self.poll_interval = poll_interval
        self._stop = threading.Event()
        self._counter = 0                # tie-breaker so callables aren't compared
        self._lock = threading.Lock()

    def schedule(self, delay, func, *args, **kwargs):
        exec_time = time.time() + delay
        with self._lock:
            count = self._counter
            self._counter += 1
        self.q.put((exec_time, count, func, args, kwargs))

    def _run(self):
        while not self._stop.is_set():
            try:
                exec_time, count, func, args, kwargs = self.q.get(timeout=self.poll_interval)
            except queue.Empty:
                continue
            now = time.time()
            if now < exec_time:
                # not ready yet -> wait the remaining time, then run
                time.sleep(exec_time - now)
            func(*args, **kwargs)
            self.q.task_done()

    def start(self):
        self.worker = threading.Thread(target=self._run, daemon=True)
        self.worker.start()

    def wait_all(self):
        self.q.join()

    def stop(self):
        self._stop.set()

def make_task(name):
    return lambda: print(f"[{time.strftime('%H:%M:%S')}] Task '{name}' executed")

sched = TaskScheduler()
sched.start()

sched.schedule(3, make_task("C (3s)"))
sched.schedule(1, make_task("A (1s)"))
sched.schedule(2, make_task("B (2s)"))

print(f"[{time.strftime('%H:%M:%S')}] All tasks scheduled")
sched.wait_all()
sched.stop()
print("All scheduled tasks completed (note they ran in delay order: A, B, C)")
````

# Problem 3: Bounded Task Queue with Multiple Producers

**Context**

A video-processing company receives video upload requests from multiple users simultaneously.

The upload requests are generated by several producer threads. These requests are placed into a shared queue and processed by worker threads.

The server has limited memory, so the queue can only hold a fixed number of tasks at any given time. If the queue becomes full, producers must wait until consumers free up space.

This simulates a real-world producer-consumer system with resource constraints.

**Queue**

Create a queue with a maximum size of 3:

```python
task_queue = queue.Queue(maxsize=3)
```

**Producers**

Create two producer threads:

- Producer-1 generates tasks 1-5.
- Producer-2 generates tasks 6-10.

Each producer should:

1. Create a task.
2. Add it to the queue.
3. Print a message indicating the task was added.
4. Sleep for 0.5 seconds before producing the next task.

Example:

```text
Producer-1 added task 1
Producer-1 added task 2
```

---

**Consumers**

Create two worker threads.

Each worker should:

1. Retrieve a task from the queue.
2. Print that processing has started.
3. Sleep for 2 seconds to simulate work.
4. Print that processing has finished.

Example:

```text
Worker-1 processing task 3
Worker-1 finished task 3
```

---

**Queue Full Behaviour**

Before inserting a task, producers should display:

```text
Producer-1 attempting to add task 4
```
If the queue is full, the producer will automatically block until a worker removes a task.

Observe the execution order carefully.

---

**Shutdown**

Once all tasks are processed:

1. Wait for the queue to become empty.
2. Send a termination signal (`None`) to each worker.
3. Wait for all workers to exit.

---

**[Example] Output**

```text
Producer-1 attempting to add task 1
Producer-1 added task 1

Producer-2 attempting to add task 6
Producer-2 added task 6

Worker-1 processing task 1
Worker-2 processing task 6

Producer-1 attempting to add task 2
Producer-1 added task 2

Producer-2 attempting to add task 7
Producer-2 added task 7

Producer-1 attempting to add task 3
Producer-1 added task 3

Producer-2 attempting to add task 8

Worker-1 finished task 1
Worker-1 processing task 2

Producer-2 added task 8

...
```

---

````python

import threading
import queue
import time

task_queue = queue.Queue(maxsize=3)
print_lock = threading.Lock()

def log(msg):
    with print_lock:
        print(msg)

def producer(name, task_ids):
    for tid in task_ids:
        log(f"{name} attempting to add task {tid}")
        task_queue.put(tid)          # blocks if queue is full
        log(f"{name} added task {tid}")
        time.sleep(0.5)

def worker(name):
    while True:
        tid = task_queue.get()
        if tid is None:              # termination signal
            task_queue.task_done()
            break
        log(f"{name} processing task {tid}")
        time.sleep(2)                # simulate work
        log(f"{name} finished task {tid}")
        task_queue.task_done()

producers = [
    threading.Thread(target=producer, args=("Producer-1", range(1, 6))),
    threading.Thread(target=producer, args=("Producer-2", range(6, 11))),
]
workers = [
    threading.Thread(target=worker, args=("Worker-1",)),
    threading.Thread(target=worker, args=("Worker-2",)),
]

for w in workers: w.start()
for p in producers: p.start()

for p in producers: p.join()

task_queue.join()

for _ in workers:
    task_queue.put(None)

for w in workers: w.join()

print("\nAll tasks processed and workers shut down cleanly.")
````
