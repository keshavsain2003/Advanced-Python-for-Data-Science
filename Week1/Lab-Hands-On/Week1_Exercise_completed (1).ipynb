## **Problem 1: Lazy Evaluation with Generators**
A generator produces values one at a time instead of storing the entire result in memory.

### Task

1. Write a generator function `count_up_to(n)` that yields numbers from `1` to `n`.
2. Use a `for` loop to print the generated values.
3. Print the type of the object returned by the generator function.

### Expected behaviour

Example:
```python
count_up_to(5)
```

Output:

1
2
3
4
5

### Reflection Questions
- Why is a generator more memory efficient than returning a full list?
- When would you design a function as a generator rather than returning a collection?
- What is the effect of an infinite loop (`while True`) inside a generator function?
- What happens if a generator contains both `yield` and `return` statements?
---

````python
def count_up_to(n):
  count = 0
  while count < n:
    yield count
    count += 1

gen = count_up_to(5)
for _ in range(5):
  print(next(gen))

# Que:
# Genrator are more memory efficient because it loads one item at a time.
# We will use generator when we are working with large datasets or when we want to create an infinite sequence of values.
# infinite loop in gernator will not cause memory issues because it generates values on-the-fly and does not store them all in memory at once.
# if generator ontains both yield and return, the return statement will raise a StopIteration exception when the generator is exhausted, and the value returned will be available as the value of the exception.  
````

## **Problem 2: Data Pipeline using Generators**

Generators are often composed to create efficient data pipelines.

Consider a list of text entries:

data = ["apple", "", "banana", "cherry", "", "date"]

### Task

Create a generator pipeline consisting of three steps:

1. `filter_empty()`  
   Remove empty strings.

2. `make_uppercase()`  
   Convert each word to uppercase.

3. `add_length()`  
   Return a tuple containing the word and its length.

The final output should look like:

```python
('APPLE', 5)
('BANANA', 6)
('CHERRY', 6)
('DATE', 4)
```
### Question
What advantage does this pipeline approach provide when working with very large datasets?

---


````python
data = ["apple", "", "banana", "cherry", "", "date"]
def filter_empty(data):
    for item in data:
        if item:
            yield item

def make_uppercase(items):
    for item in items:
        yield item.upper()

def add_length(items):
    for item in items:
        yield (item, len(item))

for word, length in add_length(make_uppercase(filter_empty(data))):
    print(f"Item: {word}, Length: {length}")

# The advantage with this pipeline is that it allows us to process data in a memory-efficient way, as each step of the pipeline generates values on-the-fly without needing to store intermediate results in memory. This is particularly beneficial when working with large datasets, as it minimizes memory usage and allows for efficient data processing.

````

## **Problem 3: Writing a Decorator**

Decorators modify the behaviour of a function without changing its source code.

### Task

1. Write a decorator called `log_calls`.
2. The decorator should print: Function <function_name> was called
3. Apply the decorator to a function `add(a, b)`.

### Expected behaviour

Calling:

add(3, 4)

Output:

Function add was called
7

### Question

Why are decorators useful in large software systems?

---


````python
from functools import wraps

def log_calls(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling function: {func.__name__} with args: {args}")
        result = func(*args, **kwargs)
        print(f"Function {func.__name__} returned: {result}")
        return result
    return wrapper

@log_calls
def add(a, b):
    return a + b
add(3, 4)

# decorator are usefly in large software development for logging, authentication, and performance monitoring. They allow us to add functionality to existing functions without modifying their code, making our code more modular and easier to maintain.
````

## **Problem 4: Counting Function Calls**
Decorators can also store state.

### Task

Create a decorator `count_calls` that keeps track of how many times a function is executed.

Each call should print:
- Call number X
- Apply the decorator to a function `greet(name)`.

### Example output

Call number 1
Hello Alice

Call number 2
Hello Bob

---

````python
class CallLog:
    def __init__(self, func):
        self.func = func
        self.count = 0

    def __call__(self, *args, **kwargs):
        self.count += 1
        print(f"Call number {self.count}")
        return self.func(*args, **kwargs)
        
@CallLog
def greet(name):
    print(f"Hello {name}")
greet('Yash')
greet('Gaurav')
greet('Parth')


````

## **Problem 5: Context Managers**

Context managers control resource acquisition and release.

### Task

Create a context manager called `managed_resource(name)` using `contextlib.contextmanager`.

The behaviour should be:

Opening <name>
Working with <name>
Closing <name>

### Example usage
```python
with managed_resource("database") as db:
    print("Working with", db)
```
### Question

Why are context managers useful for handling resources like files or database connections?

---

````python
class managed_resource:
    def __init__(self, name):
        self.name = name

    def __enter__(self):
        print(f"Opening {self.name}")
        return self.name

    def __exit__(self, exc_type, exc_val, exc_tb):
        print(f"Closing {self.name}")
with managed_resource("database") as db:
    print(f"Working with {db}")

# context manager is useful for handling db and files because it ensures that resources are properly released after use, even if an error occurs. This helps prevent resource leaks and ensures that connections are closed, which is crucial for maintaining the stability and performance of applications.
````

## **Problem 6: Performance Monitoring**


In large applications, it is useful to measure how long functions take to execute.

### Task

Create a decorator called `timer` that:

1. Measures the execution time of a function
2. Prints:
```python 
Function <name> executed in <time> seconds
```
Apply it to a function that computes the sum of numbers from 1 to 1,000,000.

### Question
Why is a decorator better than manually inserting timing code inside every function?

---

````python
import time
from functools import wraps

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start  = time.time()
        result = func(*args, **kwargs)
        end    = time.time()
        print(f"Function {func.__name__} executed in {end - start:.4f} seconds")
        return result
    return wrapper

@timer
def compute_sum(n):

    return sum(range(1, n + 1))

result = compute_sum(1_000_000)
print(f"Result: {result}")
````

Consider a system that processes a stream of sensor readings collected from multiple devices. 
Each reading is represented as a tuple `(device_id, value)`.

Some readings are invalid (negative values) and must be discarded. The remaining values must be normalised and transformed for further analysis.

readings = [
    ("A", 10),

#### Extension 3: Function Wrapping

Define a wrapper function `safe_process(readings)` that:

- raises a controlled exception otherwise

````python
from functools import wraps

readings = [
  ("A", 10),
  ("B", -3),
  ("A", 25),
  ("C", 0),
  ("B", 15),
]

data = filter(lambda x: x[1] > 0, readings)
data = map(lambda x: (x[0], x[1] * 2), data)
print(list(data))

# we use map and filter for data processing because they allow us to apply transformations and filters to data in a concise and efficient way. They are particularly useful when working with large datasets, as they can be combined with generators to process data on-the-fly without needing to store intermediate results in memory. This makes our code more efficient and easier to read.

def decorate(func):
  @wraps(func)
  def wrapper(*args, **kwargs):
    print(f"Entering {func.__name__}")
    for item in func(*args, **kwargs):
      yield item
    print(f"Exiting {func.__name__}")
  return wrapper

@decorate
def process(readings):
  for i, j in readings:
    if j > 0:
      yield (i, j * 2)

dats = process(readings)
print(list(dats))
````
